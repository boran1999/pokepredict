```python
# -*- coding: utf-8 -*-
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import tree
import pickle

ff=open("spok.dat","rb")
s2=ff.read()
ff.close()
clf=pickle.loads(s2)

f=open("pokemon.csv")
dict = {
        'Bug': 1,
        'Dark': 2,
        'Dragon': 3,
        'Electric': 4,
        'Fairy': 5,
        'Fighting': 6,
        'Fire': 7,
        'Flying': 8,
        'Ghost': 9,
        'Grass': 10,
        'Ground': 11,
        'Ice': 12,
        'Normal': 13,
        'Poison': 14,
        'Psychic': 15,
        'Rock': 16,
        'Steel': 17,
        'Water': 18}

gsarr=[]
yarr=[]
tarr=[]

for line in f:
    arr = line.split(',')
    if arr[3] == '':
        tarr.append(arr[0])
        gsarr.append(dict[arr[2]])
        x = [float(arr[6]),float(arr[7]),float(arr[8]),float(arr[10])]
        pred = clf.predict([x])
        yarr.append(float(pred[0]))
summ=0
count=0 
for i in range(len(gsarr)):
    if gsarr[i]==yarr[i]:
        summ+=1
    count+=1
print(summ/count)

#print(timearr)

#plt.plot(tarr, yarr, tarr, gsarr)
#plt.show()

```
